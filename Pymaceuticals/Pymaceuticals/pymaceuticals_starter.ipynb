```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
file_1 = "data/mouse_drug_data.csv"
file_2 = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_drug_df = pd.read_csv(file_1)
clinicaltrial_df = pd.read_csv(file_2)
# Combine the data into a single dataset
merge_df = pd.merge(mouse_drug_df, clinicaltrial_df, on="Mouse ID")

# Display the data table for preview

merge_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint
drug_timepoint_data = merge_df.groupby(["Drug", "Timepoint"])
avg_tumor_vol = drug_timepoint_data["Tumor Volume (mm3)"].mean()

# Convert to DataFrame
summary_data_avg = pd.DataFrame(avg_tumor_vol)

# Preview DataFrame
summary_data_avg
```

```python
summary_data_avg.head(100)
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
drug_timepoint_data = merge_df.groupby(["Drug", "Timepoint"])
standard_error_tumor_data = drug_timepoint_data["Tumor Volume (mm3)"].sem()

# Convert to DataFrame
summary_data_sem = pd.DataFrame(standard_error_tumor_data)

# Preview DataFrame
summary_data_sem
```

```python
# Minor Data Munging to Re-Format the Data Frames
reformat_pivot_avg = summary_data_avg.pivot_table(index = "Timepoint", columns = "Drug", values = "Tumor Volume (mm3)")
reformat_pivot_sem = summary_data_sem.pivot_table(index = "Timepoint", columns = "Drug", values = "Tumor Volume (mm3)")

# Preview that Reformatting worked
reformat_pivot_avg.head()
```

```python
# Generate the plots (with errorbars)

fig, ax = plt.subplots()

reformat_pivot_avg = summary_data_avg.pivot_table(index = "Timepoint", columns = "Drug", values = "Tumor Volume (mm3)")
reformat_pivot_sem = summary_data_sem.pivot_table(index = "Timepoint", columns = "Drug", values = "Tumor Volume (mm3)")

ax.errorbar(reformat_pivot_avg.index, reformat_pivot_avg["Capomulin"], yerr=reformat_pivot_sem["Capomulin"], 
             color="red", fmt="o", alpha=0.5, label="Tumor Volume (mm3) of Capomulin")
ax.errorbar(reformat_pivot_avg.index, reformat_pivot_avg["Infubinol"], yerr=reformat_pivot_sem["Infubinol"], 
             color="blue", fmt="^", alpha=0.5, label="Tumor Volume (mm3) of Infubinol")
ax.errorbar(reformat_pivot_avg.index, reformat_pivot_avg["Ketapril"], yerr=reformat_pivot_sem["Ketapril"], 
             color="green", fmt="-", alpha=0.5, label="Tumor Volume (mm3) of Ketapril")
ax.errorbar(reformat_pivot_avg.index, reformat_pivot_avg["Placebo"], yerr=reformat_pivot_sem["Placebo"], 
             color="black", fmt="s", alpha=0.5, label="Tumor Volume (mm3) of Placebo")

plt.legend(loc="best", fontsize="small", fancybox=True)
plt.title("Tumor Responses to Drugs")

ax.set_xlabel("Time")
ax.set_ylabel("Mean of Tumor Volume (mma)")

plt.grid()
plt.savefig("Tumor_Responses_To_Drugs.png")

plt.show()

# Save the Figure
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
drug_timepoint_data = merge_df.groupby(["Drug", "Timepoint"])
avg_metastatic_sites = drug_timepoint_data["Metastatic Sites"].mean()

# Convert to DataFrame
summary_data_ms = pd.DataFrame(avg_metastatic_sites)

# Preview DataFrame
summary_data_ms
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
drug_timepoint_data = merge_df.groupby(["Drug", "Timepoint"])
standard_error_metastatic_data = drug_timepoint_data["Metastatic Sites"].sem()

# Convert to DataFrame
summary_data_ms_sem = pd.DataFrame(standard_error_metastatic_data)

# Preview DataFrame
summary_data_ms_sem
```

```python
# Minor Data Munging to Re-Format the Data Frames
reformat_pivot_ms = summary_data_ms.pivot_table(index = "Timepoint", columns = "Drug", values = "Metastatic Sites")
reformat_pivot_ms_sem = summary_data_ms_sem.pivot_table(index = "Timepoint", columns = "Drug", values = "Metastatic Sites")

# Preview that Reformatting worked
reformat_pivot_ms_sem
```

```python
# Generate the Plot (with Error Bars)
fig, ax = plt.subplots()

reformat_pivot_ms = summary_data_ms.pivot_table(index = "Timepoint", columns = "Drug", values = "Metastatic Sites")
reformat_pivot_ms_sem  = summary_data_ms_sem.pivot_table(index = "Timepoint", columns = "Drug", values = "Metastatic Sites")

ax.errorbar(reformat_pivot_ms.index, reformat_pivot_ms["Capomulin"], yerr=reformat_pivot_ms_sem["Capomulin"], 
             color="red", fmt="o", alpha=0.5, label="Metastatic Sites of Capomulin")
ax.errorbar(reformat_pivot_ms.index, reformat_pivot_ms["Infubinol"], yerr=reformat_pivot_ms_sem["Infubinol"], 
             color="blue", fmt="^", alpha=0.5, label="Metastatic Sites of Infubinol")
ax.errorbar(reformat_pivot_ms.index, reformat_pivot_ms["Ketapril"], yerr=reformat_pivot_ms_sem["Ketapril"], 
             color="green", fmt="-", alpha=0.5, label="Metastatic Sites of Ketapril")
ax.errorbar(reformat_pivot_ms.index, reformat_pivot_ms["Placebo"], yerr=reformat_pivot_ms_sem["Placebo"], 
             color="black", fmt="s", alpha=0.5, label="Metastatic Sites of Placebo")

plt.legend(loc="best", fontsize="small", fancybox=True)
plt.title("Metastatic Sites of Drugs")

ax.set_xlabel("Time")
ax.set_ylabel("Metastatic Sites")

plt.grid()

# Save the Figure
plt.savefig("Images/MetastaticSitesofDrugs.png")

# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
drug_timepoint_data = merge_df.groupby(["Drug", "Timepoint"])
mouse_count = drug_timepoint_data["Mouse ID"].count()

# Convert to DataFrame
summary_data_sr = pd.DataFrame(mouse_count)

# Preview DataFrame
summary_data_sr
```

```python
# Minor Data Munging to Re-Format the Data Frames
reformat_pivot_sr = summary_data_sr.pivot_table(index = "Timepoint", columns = "Drug", values = "Mouse ID")

# Preview the Data Frame
reformat_pivot_sr
```

```python
# Generate the Plot (Accounting for percentages)
fig, ax = plt.subplots()

reformat_pivot_sr = summary_data_sr.pivot_table(index = "Timepoint", columns = "Drug", values = "Mouse ID")

ax.errorbar(reformat_pivot_sr.index, (reformat_pivot_sr["Capomulin"] / 25) * 100, color="red", fmt="o", alpha=0.5, label="Mouse IDs of Capomulin")
ax.errorbar(reformat_pivot_sr.index, (reformat_pivot_sr["Infubinol"] / 25) * 100, color="blue", fmt="^", alpha=0.5, label="Mouse IDs of Infubinol")
ax.errorbar(reformat_pivot_sr.index, (reformat_pivot_sr["Ketapril"] / 25) * 100, color="green", fmt="-", alpha=0.5, label="Mouse IDs of Ketapril")
ax.errorbar(reformat_pivot_sr.index, (reformat_pivot_sr["Placebo"] / 25) * 100, color="black", fmt="s", alpha=0.5, label="Mouse IDs of Placebo")

plt.legend(loc="best", fontsize="small", fancybox=True)
plt.title("Survival of Mice ")

ax.set_xlabel("Time")
ax.set_ylabel("Survival Percentage")

plt.grid()

# Save the Figure
plt.savefig("Images/Survival.png")

# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
timepoint_0_tumor_vol = 45
reformat_pivot_avg = summary_data_avg.pivot_table(index = "Timepoint", columns = "Drug", values = "Tumor Volume (mm3)")
percent_change_per_drug = ((reformat_pivot_avg.loc[45, :] - timepoint_0_tumor_vol)/timepoint_0_tumor_vol)*100
percent_change_per_drug = round(percent_change_per_drug, 2)

# Display the data to confirm
percent_change_per_drug
```

```python
# Store all Relevant Percent Changes into a Tuple
percent_changes_to_tuple = (percent_change_per_drug["Capomulin"], 
               percent_change_per_drug["Infubinol"], 
               percent_change_per_drug["Ketapril"], 
               percent_change_per_drug["Placebo"])

# Splice the data between passing and failing drugs
fig, ax = plt.subplots()

x_axis = np.arange(len(percent_changes_to_tuple))
width = 1
passed = ax.bar(x_axis[0], percent_changes_to_tuple[0], width, color='green')
failed = ax.bar(x_axis[1:], percent_changes_to_tuple[1:], width, color='red')

# Wider width. Add labels, tick marks, etc. 
plt.title('Tumor Change Over 45 Days')
plt.legend(loc="best", fontsize="small", fancybox=True)

ax.set_xlabel ("Drug")
ax.set_ylabel("Percentage of Tumor Volume Change")

ax.set_xticks(x_axis + 0.5)
ax.set_xticklabels(("Capomulin", "Infubinol", "Ketapril", "Placebo"))
plt.xticks(ha="right")

ax.set_ylim([-25,70])
ax.grid()


# Use functions to label the percentages of changes
def autolabelfailed(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate("{}".format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha="center", va="bottom", color="black")
def autolabelpassed(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate("{}".format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),
                    textcoords="offset points",
                    ha="center", va="bottom", color="black")

# Call functions to implement the function calls
autolabelfailed(failed)
autolabelpassed(passed)

# Save the Figure
plt.savefig("Images/Tumor_Volume_Change_Percentage.png")

# Show the Figure
plt.show()
fig.show()
```
